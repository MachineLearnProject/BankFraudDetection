```python
import pandas as pd
import numpy as np
```

```python
#read in CSV 

df = pd.read_csv("PS_20174392719_1491204439457_log.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()

df.head()
```

```python
# See if what the net balance is to understand data
df['netbalanceOrig'] = df['oldbalanceOrg'] - df['newbalanceOrig']
df['netbalanceDest'] = df['oldbalanceDest'] - df['newbalanceDest']
df
```

```python
#Understand what are the high low values to determine how to analyze data.  Specifically to see what is the highest transaction amounts.

df.describe()
```

```python
df.groupby(["isFraud", "type"])["amount"].sum()
```

```python
#size of data set both rows and columns
df.shape
```

```python
 #Understand which values are merchant and which ones are customer as customer ones most likely to be fraudulent

def check_string(input):
    if input.startswith("M"):
        return 0
    else:
        return 1

df["M or C"] = df["nameDest"].apply(lambda x:check_string(x))
df
```

```python
#Determine if merchant or client data
df["M or C"].value_counts()
```

```python

#Accounts that are emptied of all their funds are predominantly fraudulent in this data set.  So we are looking for zero values
df["ZeroBal"]= np.where(df["newbalanceOrig"]==0, 1, 0)
#Transactions that are greater than P75 or $200000 are also flagged as potential frauds
df["LargeTrans"]= np.where(df["newbalanceOrig"]>=200000, 1, 0)

df
```

```python
#Removed columns that were used to calculate values below.
new_df = df.drop(columns =[  "step", "amount","nameOrig","oldbalanceOrg", "newbalanceOrig","nameDest","oldbalanceDest","newbalanceDest","isFlaggedFraud", "netbalanceOrig", "netbalanceDest"])
new_df
```

```python
#use one-hot-encoding to view transaction types with 1 or 0
clean_df =pd.get_dummies(new_df, prefix=["type"])
clean_df
```

```python
!pip install -U imbalanced-learn
from imblearn.over_sampling import SMOTE
```

```python
#use all columns in clean_df except "isFraud" to see if can calculate fraud
X = clean_df.drop(["isFraud"], axis=1)
y = clean_df["isFraud"]
print(X.shape, y.shape)
```

```python
#as we have limited fraud data oversampling is used to get additional data point in the testing dataset to quantify validity of models

oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#use 75% of data to train model and 25% to test
X_train, X_test, y_train, y_test = train_test_split(
    X, y)
```

```python
# import joblib 
# model = joblib.load('BankFraudDetection-check.sav')
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
#run and fit logistic regression model

model = GaussianNB()
model.fit(X_train,y_train)
```

```python
#save model data
import joblib
filename = 'BankFraudDetection-Naive-Bayes.sav'
joblib.dump(model, filename)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score
```

```python
print(f"Training Data Score: {precision_score(y_train, model.predict(X_train))}")
print(f"Testing Data Score: {precision_score(y_test, model.predict(X_test))}")
```

```python
print(f"Training Data Score: {recall_score(y_train, model.predict(X_train))}")
print(f"Testing Data Score: {recall_score(y_test, model.predict(X_test))}")
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score
print(f"Training Data Score: {f1_score(y_train, model.predict(X_train))}")
print(f"Testing Data Score: {f1_score(y_test, model.predict(X_test))}")
```

```python
print(f"Training Data Score: {accuracy_score(y_train, model.predict(X_train))}")
print(f"Testing Data Score: {accuracy_score(y_test, model.predict(X_test))}")
```
